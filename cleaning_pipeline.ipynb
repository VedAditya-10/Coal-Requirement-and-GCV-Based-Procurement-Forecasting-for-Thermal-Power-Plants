```python
def find_header_row(df):
    for i in range(len(df)):
        if df.iloc[i].astype(str).str.contains("Name of TPS", case=False).any():
            return i
    raise ValueError("Header row not found")
```

```python
def normalize_columns(df):
    df.columns = [
        unicodedata.normalize("NFKC", str(c))
        .replace("\n", " ")
        .replace("\xa0", " ")
        .strip()
        for c in df.columns
    ]
    return df
```

```python
def is_real_plant(name):
    if pd.isna(name):
        return False

    name = str(name).strip().lower()

    bad_patterns = [
        r"^total",
        r"^northern",
        r"^southern",
        r"^eastern",
        r"^western",
        r"^central"
    ]

    return not any(re.match(p, name) for p in bad_patterns)
```

```python
def standardize_plant_name(name):
    return (
        str(name)
        .lower()
        .strip()
        .replace("&", "and")
        .replace("-", " ")
        .replace(",", "")
    )
```

```python
def process_monthly_file(file_path, year, month, reference_columns):
    # read raw
    raw = pd.read_excel(file_path, header=None)

    # detect header
    header_row = find_header_row(raw)

    # read table
    df = pd.read_excel(file_path, header=header_row)

    # normalize columns
    df = normalize_columns(df)

    # keep real plants
    df = df[df["Name of TPS"].apply(is_real_plant)]

    # numeric coercion
    for col in df.columns:
        if col != "Name of TPS":
            df[col] = pd.to_numeric(df[col], errors="coerce")

    # add time info
    df["Year"] = year
    df["Month"] = month

    # plant key
    df["plant_key"] = df["Name of TPS"].apply(standardize_plant_name)

    # align schema
    df = df.reindex(columns=reference_columns)

    # remove zero / NaN consumption
    if "Total Consumption" in df.columns:
        df = df[df["Total Consumption"] > 0]

    return df

```
